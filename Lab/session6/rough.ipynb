## Imports and helper functions

```python
import numpy as np
import matplotlib.pyplot as plt

print('Loaded numpy and matplotlib')
```

## Visualize gradient descent path for f(x) = (x-3)^2

```python
def f(x):
    return (x-3)**2

def grad(x):
    return 2*(x-3)

w = 0.0
lr = 0.1
ws = [w]
for _ in range(20):
    w = w - lr * grad(w)
    ws.append(w)

xs = np.linspace(-5, 7, 200)
plt.plot(xs, f(xs), label='f(x)')
plt.scatter(ws, [f(w) for w in ws], color='red')
plt.title('Gradient descent path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

## SGD vs Batch GD hint (toy example)

```python
# Toy comparison: batch vs mini-batch behaviors (conceptual)
# Not a rigorous experiment; use for intuition
X = np.linspace(1,4,4).reshape(-1,1)
y = 2*X.flatten()

# Batch GD update
W = 0.0
b = 0.0
lr = 0.01
for epoch in range(5):
    y_pred = W*X.flatten() + b
    error = y_pred - y
    dW = np.mean(2*X.flatten()*error)
    db = np.mean(2*error)
    W -= lr * dW
    b -= lr * db

print('Batch GD W,b:', W, b)

# SGD (single-sample updates)
W = 0.0; b = 0.0
for epoch in range(5):
    for i in range(len(X)):
        xi = X[i]
        yi = y[i]
        y_pred = W*xi + b
        error = y_pred - yi
        dW = 2 * xi * error
        db = 2 * error
        W -= lr * dW
        b -= lr * db
print('SGD W,b:', W, b)
```
